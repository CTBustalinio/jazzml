Job # 1: find a way to convert these lens and/or offsets into durations suitable for <em>mingus</em>. Notice that if 4 is the value of a quarter note, and 8 is the value of an eighth note, then it's inversely proportional. This suggests that you may need to feed in the reciprocal of the length (or offset), perhaps:

$f(\frac{1}{x})=k*\frac{1}{\frac{1}{x}}$

where k is some constant. For example, our first note is $D5$ with a duration of $0.75$, or $3/4$. Therefore, if you want to get this in <em>mingus</em>, to get $\frac{3}{4}$, you should feed in $\frac{4}{3}$ to get $\frac{1}{4/3}$, which is therefore $\frac{1}{1/x}$.

```python
# Given a duration (say 2 for 2 quarter notes), get
# the duration in mingus form. This means that you
# can feed in the duration into mingus like
# fluidsynth.place_note(note, duration)
def toMingusBeat(duration):
    return (1. / duration)

map(toMingusBeat, oscar2["Len"])
```
