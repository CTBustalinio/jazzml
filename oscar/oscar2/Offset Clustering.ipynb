Actually, it might be useful to plot stuff first - that way you get an idea of how much notes are separated from one another.

For the offsets, you'll assume that notes only come at 0.25, 0.50, etc. instead of weird numbers like 0.682 or 1.537. This is from rounding the numbers in the N-Gram notebook.

```python
%matplotlib inline

from collections import Counter, defaultdict
from sklearn.cluster import KMeans
from itertools import izip
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sys, copy
```

```python
# Read in the data and generate the offsets.
# It's okay to generate offsets here since are trigram notes.
# Recall: len = how long note lasts, offset = when it's hit (e.g. stopwatch).
numberofitems = 50
data = pd.read_csv('./oscar2trigrams.txt', names=['Note','Len'])[:numberofitems]
totaloffset = 0
offsets = []
for i in data['Len']:
    offsets.append(totaloffset)
    totaloffset += i
data["Offset"] = pd.Series(offsets)
print "Shape: (%s,%s)" % (data.shape)
data.head()
```

Next, let's see what our data looks like. For each note, we'll plot len (y axis) over offset (x axis).

```python
dx = data['Offset']
dy = data['Len']
print len(dx), len(dy)
plt.plot(dx, dy, 'm.--', linewidth=1.5)
for x, y in zip(dx, dy):
    plt.plot(x, y, 'ko')
plt.ylim([0, max(dy) + 0.5])
plt.xlim([-1, max(dx) + 0.5])
plt.grid()

# plot the ticks
# xticks = range(0, numberofitems)
# plt.xticks(xticks, data['Note'])
plt.xticks(range(0, int(max(data['Offset'])) + 1))

# Annotate with the note data
for note, offset, length in izip(data['Note'], dx, dy):
    plt.annotate(note, xy=(offset, length), color='m')

# set fig size and display
fig = plt.gcf()
fig.set_size_inches(18, 6)
plt.show()
```

It looks like we've got quite a few clusters! But be careful: those outliers with a higher length, for example, should really go with the previous notes (since length is the value that comes after).
