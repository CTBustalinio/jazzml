This is a notebook to play the original alice MIDI note information extracted into a .txt file by alice.py.

```python
import sys, time, re
import pandas as pd
import numpy as np
import music21
import mingus.core.notes as notes
sys.path.append('/usr/local/lib/python2.7/dist-packages/fluidsynth/')
from mingus.midi import fluidsynth
from mingus.containers.Note import Note
from mingus.containers.Bar import Bar
fluidsynth.init('/usr/share/sounds/sf2/FluidR3_GM.sf2',"alsa")
```

```python
# Import and order the original notes.
alice = pd.read_csv('alicenotes.txt', skiprows=2)[:].sort("Offset")
alice.index = xrange(1, len(alice) + 1)
alice = alice[alice.Octave >= 4] # threshold >= octave 4 for melodies
with open('alicenotes.txt', 'rb') as f:
    metmark = float(f.readline())
    tsig_num, tsig_den = [i for i in f.readline().replace(' /', '').split()]
    
print "Metrics:" 
print metmark, tsig_num, tsig_den, len(alice) # len shorter if octave cutoff, also if [:200] in .read_csv
alice.head(20)
```

```python
# Convert music21 note to mingus note. Takes string, returns Note.
def mingify(note):
    accidental = re.compile("[A-Z](-|#)[0-9]")
    if accidental.match(note):
        if '-' not in note:
            note = Note("%s%s-%s" % (note[0], note[1], note[2]))
        else:
            note = Note(note.replace('-', 'b-'))
    else:
        note = Note("%s-%s" % (note[0], note[1]))
    return note
    
# Get the notes. Note that you need to convert "E-" to "Eb" since you're converting music21 to mingus.
allnotes = []
alldurations = []
for note, octave in zip(alice["Note/Rest"], alice["Octave"]):
    allnotes.append(mingify("%s%s" % (note, octave)))
for length in alice["Len"]:
    alldurations.append(length)
```

```python
# pop everything into one phat bar
b = Bar()
b.set_meter((sum(alldurations), 4))
for ix, note in enumerate(allnotes):
#     print note, alldurations[ix]
#     b.place_notes(note, alldurations[ix])
    b.place_notes(note, 1)
```

```python
fluidsynth.play_Bar(b, 1, 500)
```
